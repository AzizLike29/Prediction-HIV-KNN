```python
from sklearn.preprocessing import LabelEncoder
import pandas as pd

# Membaca data dari file CSV asli
data = pd.read_csv('Laporan_HIV.csv')

# Kolom kategorikal yang akan diubah menjadi numerik
categorical_cols = ['Umur', 'Jenis Kelamin', 'Kelompok Populasi', 'Alasan Kunjungan', 'Status ODHIV']

# Dictionary untuk menyimpan label asli dan angka yang dihasilkan
label_mappings = {}

# Inisialisasi LabelEncoder dan melakukan encoding untuk setiap kolom
for col in categorical_cols:
    label_encoder = LabelEncoder()
    data[col] = label_encoder.fit_transform(data[col])

    # Simpan label asli beserta encoding numeriknya
    label_mappings[col] = dict(zip(label_encoder.transform(label_encoder.classes_), label_encoder.classes_))

# Tampilkan data setelah label encoding
print("\nData setelah label encoding:")
print(data)

# Tampilkan hasil mapping label dan angka dalam bentuk tabel yang rapi
print("\nMapping label untuk setiap kolom kategorikal dengan penjelasan:")

# Membuat tabel dari hasil mapping
for col, mapping in label_mappings.items():
    print(f"\n{col}:")
    df_mapping = pd.DataFrame(list(mapping.items()), columns=['Angka', 'Label'])
    print(df_mapping.to_string(index=False))
```
